```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import sklearn
import numpy as np
from sklearn.preprocessing import normalize
import seaborn as sns
import matplotlib.pyplot as plt

df_train_pre = pd.read_csv("fraudTrain.csv")
df_test_pre = pd.read_csv("fraudTest.csv")

df_train_pre.drop(df_train_pre.columns[0],axis=1, inplace=True)
df_test_pre.drop(df_test_pre.columns[0],axis=1,inplace=True)
df_train_pre = df_train_pre[['cc_num','amt','lat','long','merch_lat','merch_long','is_fraud']]
df_test_pre = df_test_pre[['cc_num','amt','lat','long','merch_lat','merch_long','is_fraud']]
```

```python
rows = 100000
df_train = df_train_pre.sort_values(by=['is_fraud'], ascending=False).head(rows)
print(df_train)
x_train=df_train.drop("is_fraud",axis=1)
y_train=df_train["is_fraud"]
x_test=df_test_pre.drop("is_fraud",axis=1)
y_test=df_test_pre["is_fraud"]
```

```python
print("Building training and testing datasets")
print()

print("train data shape",x_train.shape)
print("test data shape",x_test.shape)
print("y train shape", y_train.shape)
print("y test shape", y_test.shape)

print("Training RandomForest Model....")
print()
model = RandomForestClassifier(n_estimators=300, random_state=42,class_weight='balanced',verbose=2)

model.fit(x_train.values,y_train.values)
print()

print()
y_pred_test = model.predict(x_test.values)
test_acc = accuracy_score(y_test, y_pred_test)
test_rep = classification_report(y_test, y_pred_test)
conf_mat = confusion_matrix(y_test.values, y_pred_test)
norm_mat = normalize(conf_mat, axis=1, norm='l1')

print()
print("---- METRICS RESULTS FOR TESTING DATA ----")
print()
print("Total Rows are:", x_test.shape[0])
print("[TESTING] Model Accuracy is:", test_acc)
print("[TESTING] Confusion Matrix:")
print(conf_mat)
print("[TESTING] Normalized Confusion Matrix:")
print(norm_mat)
print("[TESTING] Testing Report:")
print(test_rep)

plt.figure(figsize=(10, 7))
sns.heatmap(norm_mat, annot=True, fmt='.2f', cmap='Blues', xticklabels=['Not fraud', 'Fraud'], yticklabels=['Not fraud', 'Fraud'])
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title(f'Normalized Confusion Matrix\nAccuracy: {test_acc:.4f}, Training Data: {rows}')

# Save the heatmap to a file
# plt.savefig(f'hmap{rows}.png')
plt.show()
```
